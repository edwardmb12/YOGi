## Processing image sizes

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
folder_list = os.listdir("/Users/ejm-b/code/edwardmb12/YOGi/Yoga_test_data/")
folder_list.remove('.DS_Store')
```

```python
file_path_list =[]
for folder in folder_list:
    folder_path = f"/Users/ejm-b/code/edwardmb12/YOGi/Yoga_test_data/{folder}/"
    file_path=os.listdir(folder_path)

    for file in file_path:
        if file == '.DS_Store':
            pass
        else:
            file_path = (f"/Users/ejm-b/code/edwardmb12/YOGi/Yoga_test_data/{folder}/{file}")
            file_path_list.append(file_path)
```

```python
image_list = [cv2.imread(file_path) for file_path in file_path_list]
```

```python
image_sizes = [image_list[i].shape for i in range(len(image_list))]    
```

```python
landscape_images = []
portrait_images = []
for image_size in image_sizes:
    if image_size[0] < image_size[1]:
               landscape_images.append(image_size)
    else:
        portrait_images.append(image_size)
```

```python
landscape_images = np.array(landscape_images)
portrait_images = np.array(portrait_images)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 5))
sns.histplot(landscape_images[:,0], bins = 30, ax= axes[0]).set(title='Landscape: Heights', xlabel='Height')
sns.histplot(landscape_images[:,1], bins = 30, ax=axes[1]).set(title='Landscape: Widths', xlabel='Width')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 5))
sns.histplot(portrait_images[:,0], bins = 30, ax= axes[0]).set(title='Portrait: Heights', xlabel='Height')
sns.histplot(portrait_images[:,1], bins = 30, ax=axes[1]).set(title='Portrait: Widths', xlabel='Width')
plt.show()
```

```python
print(f"LANDSCAPE: mean height {landscape_images[:,0].mean()}, mean width {landscape_images[:,1].mean()}")
```

```python
print(f"PORTRAIT: mean height {portrait_images[:,0].mean()}, mean width {portrait_images[:,1].mean()}")
```
